```python
from models import utils, metrics, plots, pipeline
from models.classifiers.Perceptron import *

from sklearn.datasets import load_iris
import pandas as pd
import numpy as np
import random
import seaborn as sns
```

```python
data = load_iris()
df_iris = pd.DataFrame(data.data, columns=data.feature_names)
df_iris['target'] = pd.Series(data.target)
```

```python
dict_df = utils.target_one_vs_all(df_iris, col_name_target='target') # Pegando dataframes por classe, exemplo: setosa vs outras

df_iris_setosa = dict_df[0] # setosa o target é 0
df_iris_versicolor = dict_df[1] # setosa o target é 1
df_iris_virginica = dict_df[2] # setosa o target é 2
```

# Setosa vs outras

```python
pipe = pipeline.make_pipeline(n_realizations = 20)

acc_setosa, desvio_padrao_setosa, realization_select_setosa, dict_realizations_setosa = pipe.run_realizations(
    
    df=df_iris_setosa, 
    k_fold=5, 
    param_grid={'learning_rate': [0.5, 0.1, 0.2]}, 
    col_target='target', 
    train_size=0.8, 
    stratify=True, 
    n_epochs=10,
    normalize=True
)
```

```python
best_perceptron_setosa = realization_select_setosa['classifier']
best_train_data = realization_select_setosa['train_data']
best_train_data = realization_select_setosa['test_data']
taxa_acerto = realization_select_setosa['taxa_acerto']
confusion_matrix = realization_select_setosa['confusion_matrix']
```

```python
print(f'Setosa\nAcurácia: {acc_setosa}\nDesvio Padrao: {desvio_padrao_setosa}')

print('\n- Info best model setosa - ')
print(f'Taxa Acerto: {taxa_acerto}')
print(f'Matriz confusão: ')
print(confusion_matrix)
```

# Criando dataframe artificial

Como solicitado no trabalho, será criado um dataframe artificial desbalanceado

```python
conjunto_1 = utils.make_coords(coord_limit_x = (-2, 2), 
                               coord_limit_y = (4, 8),
                               target= 0,
                               len_set = 10) # Top left

conjunto_2 = utils.make_coords(coord_limit_x = (-2, 2), 
                               coord_limit_y = (-2, 2),
                               target= 0,
                               len_set = 10) # Botton left

conjunto_3 = utils.make_coords(coord_limit_x = (4, 8), 
                               coord_limit_y = (-2, 2),
                               target= 0,
                               len_set = 10) # Bottom right

conjunto_4 = utils.make_coords(coord_limit_x = (4, 8), 
                               coord_limit_y = (4, 8),
                               target= 1,
                               len_set = 10) # Top right

df_artificial_1 = pd.DataFrame(data= conjunto_1 + conjunto_2 + conjunto_3 + conjunto_4, columns=['x', 'y', 'target'])
```

```python
best_perceptron_artificial = realization_select_artificial['classifier']
best_train_data = realization_select_artificial['train_data']
best_train_data = realization_select_artificial['test_data']
taxa_acerto = realization_select_artificial['taxa_acerto']
confusion_matrix = realization_select_artificial['confusion_matrix']
```

```python
print(f'Artificial\nAcurácia: {acc_artificial}\nDesvio Padrao: {desvio_padrao_artificial}')

print('\n- Info best model virginica - ')
print(f'Taxa Acerto: {taxa_acerto}')
print(f'Matriz confusão: ')
print(confusion_matrix)
```
