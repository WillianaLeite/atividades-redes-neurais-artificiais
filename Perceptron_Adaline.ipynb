# Artificial I

```python
# Com ruído

def f(x, a, b):
    return a*x + b #f(x) = ax+b 

a, b = 2, 2
x = list(range(-10, 10))
y = [f(i, a, b) + random.uniform(0,6) for i in x]

df_artificial_1_com_ruido =  pd.DataFrame(data=zip(x,y), columns =['x', 'target'])

df_artificial_1_com_ruido = utils.normalize_col(df_artificial_1_com_ruido, 'x')

df_artificial_1_com_ruido = utils.normalize_col(df_artificial_1_com_ruido, 'target')

# Sem ruído
y = [f(i, a, b) for i in x]

df_artificial_1_sem_ruido =  pd.DataFrame(data=zip(x,y), columns =['x', 'target'])

df_artificial_1_sem_ruido = utils.normalize_col(df_artificial_1_sem_ruido, 'x')

df_artificial_1_sem_ruido = utils.normalize_col(df_artificial_1_sem_ruido, 'target')
```

## Com ruído

```python
plt.scatter(df_artificial_1_com_ruido['x'], df_artificial_1_com_ruido['target'])
plt.title("Artificial I - Com ruído")
plt.xlabel("x")
plt.ylabel("y")
plt.savefig('artificial_1_com_ruido.png')
```

```python
pipe = pipelineRegressor.make_pipeline(n_realizations = 20)

mean_mse_1, mean_rmse_1, desvio_padrao_mse_1, desvio_padrao_rmse_1, dict_realizations_1 = pipe.run_realizations(
    
    df=df_artificial_1_com_ruido, 
    k_fold=5, 
    param_grid={'learning_rate': [0.0001, 0.001, 0.005]}, 
    col_target='target', 
    train_size=0.8, 
    stratify=None, 
    n_epochs=200,
    normalize=False
)
```

```python
print(f'MSE: {mean_mse_1}')
print(f'Desvio Padrão MSE: {desvio_padrao_mse_1}')
print(f'\nRMSE: {mean_rmse_1}')
print(f'Desvio Padrão RMSE: {desvio_padrao_rmse_1}')
```

### Superficie de Decisão

```python
# Selecionando a realização que teve menor MSE

menor_error = np.inf
num_best_realization = np.inf

for n_realization in dict_realizations_1.keys():
    if dict_realizations_1[n_realization]['mse'] < menor_error:
        menor_error = dict_realizations_1[n_realization]['mse']
        num_best_realization = n_realization
        
best_realization = dict_realizations_1[num_best_realization]

best_regressor_1 = best_realization['regressor']
df_train = best_realization['train_data'].drop(['x0'], axis=1)
df_test = best_realization['test_data'].drop(['x0','predict'], axis=1)

df_superficie_1 = pd.concat([df_train, df_test]).reset_index(drop=True)

plots.plot_decision_boundary(
    
    df_superficie_1,
    list_features = ['x'],
    col_target='target',
    model = best_regressor_1,
    h=.001,
    task='regressor'
)

plt.savefig('superficie_decisao_1_com_ruido.png')
```

# Artificial II

```python
x1 = list(range(1, 20))

x2 = list(range(-10, 11))

df_artificial_2_com_ruido = pd.DataFrame(data=zip(x1,x2), columns =['x1', 'x2'])

df_artificial_2_com_ruido = utils.normalize_col(df_artificial_2_com_ruido, 'x1')
df_artificial_2_com_ruido = utils.normalize_col(df_artificial_2_com_ruido, 'x2')

df_artificial_2_sem_ruido = df_artificial_2_com_ruido.copy()

a, b, c = 1, 1, 0000.1

# Com ruído
df_artificial_2_com_ruido['target'] = df_artificial_2_com_ruido.apply(lambda row: a*row['x1'] + b*row['x2'] + c + random.uniform(0,1), axis=1) 

# Sem ruído
df_artificial_2_sem_ruido['target'] = df_artificial_2_sem_ruido.apply(lambda row: a*row['x1'] + b*row['x2'] + c, axis=1) 
```

## Com ruído

```python
threedee = plt.figure(figsize=(15, 12)).gca(projection='3d')
threedee.scatter(df_artificial_2_com_ruido['x1'], df_artificial_2_com_ruido['x2'], df_artificial_2_com_ruido['target'])
threedee.set_xlabel('x1')
threedee.set_ylabel('x2')
threedee.set_zlabel('target')
plt.title("Artificial II - Com ruído")
plt.savefig('artificial_2_com_ruido.png')
plt.show()
```
